Q3, Calculate the median duration (in days) of significant market corrections in the S&P 500 index.

A3,
0.25 quantile is:  21 days
0.5 quantile is:  40 days
0.75 quantile is:  88 days

```python
SPX = yf.download('^GSPC', start="1950-01-01")
```

```python
spx_close = SPX['Close'].reset_index()
spx_close = spx_close.rename(columns={'Ticker': 'index', '^GSPC': 'Close'})
spx_close['all_time_high'] = spx_close['Close'].cummax()
spx_close['all_time_high_date'] = spx_close.apply(lambda row: row['Date'] if row['Close'] >= row['all_time_high'] else np.nan, axis=1)
spx_close['all_time_high_date'] = spx_close['all_time_high_date'].ffill()
```

```python
spx_dropdown_index = spx_close.groupby('all_time_high_date')['Close'].idxmin()
spx_dropdown = spx_close.loc[spx_dropdown_index]
spx_dropdown['dropdown_pct'] = spx_dropdown.apply(lambda row: (row['all_time_high'] - row['Close']) / row['all_time_high'], axis=1)
correction = spx_dropdown[spx_dropdown['dropdown_pct']>0.05]
correction['duration'] = correction['Date'] - correction['all_time_high_date']
correction.sort_values(by='dropdown_pct', ascending=False)
```

```python
for q in [0.25, 0.5, 0.75]:
  print(f'{q} quantile is: ', correction['duration'].quantile(q))
```

Q4, Calculate the median 2-day percentage change in stock prices following positive earnings surprises days.

A4, 2.57%

```python
amazon_url = 'https://raw.githubusercontent.com/DataTalksClub/stock-markets-analytics-zoomcamp/refs/heads/main/cohorts/2025/ha1_Amazon.csv'
amazon_eps = pd.read_csv(amazon_url, sep=';', header=0)
```

```python
ticker = yf.Ticker("AMZN")

earnings = ticker.get_earnings_dates(limit=100)
earnings = earnings.reset_index()
earnings.rename(columns={'Earnings Date': 'Date'}, inplace=True)
earnings['release_date'] = earnings['Date'].dt.tz_localize(None).dt.date
earnings
```

```python
amazon_stock_data = yf.download('AMZN')
amazon_price = amazon_stock_data['Close'].reset_index()
amazon_price = amazon_price.rename(columns={'Ticker': 'index', 'AMZN': 'Close'})
amazon_price['price_shift_2d'] = amazon_price['Close'].shift(2)
amazon_price['2d_return'] = (amazon_price['Close'] - amazon_price['price_shift_2d']) / amazon_price['price_shift_2d'] * 100
amazon_price['release_date'] = amazon_price['Date'].dt.date.shift(1)
amazon_earning_price = pd.merge(amazon_price, earnings, on='release_date', how='inner')
# amazon_earning_price = amazon_price[amazon_price['is_earning']==1]
amazon_earning_price.sort_values(by='release_date', ascending=False)
return_eps_corr = amazon_earning_price['2d_return'].corr(amazon_earning_price['Surprise(%)'])
return_eps_corr
```

```python
amazon_earning_price[amazon_earning_price['Surprise(%)']>0]['2d_return'].quantile(0.5)
```
